```python
import importlib
import tiktoken
print("Tiktoken version: ", importlib.metadata.version('tiktoken'))
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = (
    "Hello, this is a test for byte pair encoder using tiktoken. <|endoftext|>. Let's check it out."
    " Should weStartIt?"
)

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})
print(integers)
```

```python
strings = tokenizer.decode(integers)
print(strings)
```
